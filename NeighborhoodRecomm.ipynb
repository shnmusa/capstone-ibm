# Neighborhood Recommendation between Manhattan and Toronto

```python
import pandas as pd
import numpy as np

toronto = pd.read_csv("toronto.csv").drop('Unnamed: 0', 1).set_index('Neighborhood')
manhattan = pd.read_csv("manhattan.csv").drop('Unnamed: 0', 1).set_index('Neighborhood')
```

## Getting mutual venue categories

```python
venueListTor = list(toronto.columns)
venueListMan = list(manhattan.columns)
```

```python
print("Venues categories that are not included at Toronto: " + str(len(set(venueListMan) - set(venueListTor))))
print("Venues categories that are not included at Manhattan: " + str(len(set(venueListTor) - set(venueListMan))))
print("Venues categories that are included at both city: " + str(len(set(venueListMan) & set(venueListTor))))
```

```python
mutualManhattan = manhattan[set(venueListMan) & set(venueListTor)]
mutualToronto = toronto[set(venueListMan) & set(venueListTor)]
mutualToronto.head()
```

```python
mutualManhattan.head()
```

## Creating similarity matrix

```python
resultMatrix = pd.DataFrame(np.dot(mutualToronto, mutualManhattan.T))
resultMatrix.head()
```

```python
resultMatrix['Torontoâ†’'] = toronto.index.values
resultMatrix = resultMatrix.set_index('Torontoâ†’').T
resultMatrix['Manhattanâ†“'] = manhattan.index.values
resultMatrix = resultMatrix.set_index('Manhattanâ†“')
resultMatrix
```

## Getting most similar 3 neighborhood pair

```python
resultMatrix.stack().index[np.argmax(resultMatrix.values)]
```
